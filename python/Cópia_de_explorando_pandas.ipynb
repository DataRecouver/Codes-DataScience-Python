## Mesclando dataframes

### 01. Importar biblioteca

```
import numpy as np
import pandas as pd
```

### 02. Criar os dataframes data1, data2 e data3 baseados nos dados abaixo

```
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```
data1 = pd.DataFrame.from_dict(raw_data_1)
data2 = pd.DataFrame.from_dict(raw_data_2)
data3 = pd.DataFrame.from_dict(raw_data_3)
```

### 03. Crie um novo dataframe fruto da junção das linhas de data1 e data2 e armazene na variável 'all_data'

```
all_data = pd.concat([data1, data2])
all_data
```

Observe que o índice dos dois dataframes foram mantidos. Contudo, a busca por índice fica prejudicada. Selecione a linha de índice 4 e observe que apenas a primeira ocorrência é exibida.

```
print(all_data.loc[4])
print("")
print(all_data.iloc[4])
```

### 04. Resetar o índice de 'all_data'

```
all_data.reset_index()
```

### 05. Crie um outro dataframe fruto da junção das colunas de data1 e data2 e armazene na variável 'all_data_col'.

```
all_data_col = pd.concat([data1, data2], axis=1,)
all_data_col
```

### 06. Mescle os dataframes all_data e data3 com base na coluna 'subject_id'. Dica: pesquise a função merge da biblioteca Pandas

```
pd.merge(all_data,data3,on="subject_id")
```

### 07. Mescle apenas os dados que têm o mesmo 'subject_id' em data1 e data2. Dica: pesquise a função merge da biblioteca Pandas

```
pd.merge(data1,data2,on="subject_id")
```

### 08. Mescle todos os valores em data1 e data2 a partir de 'subject_id', com registros correspondentes de ambos os lados, quando disponíveis. Dica: pesquise a função merge da biblioteca Pandas

```
pd.merge(data1,data2,on="subject_id",how="outer")
```
